```python
def plot_core_with_labels(df):
    corr = df.corr()
    flg, ax = plt.subplots(figsize=(20,20))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)
```

```python
plot_core_with_labels(df)
```
